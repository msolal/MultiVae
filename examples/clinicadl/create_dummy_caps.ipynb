```python
import os
import torch
import pandas as pd 
from pathlib import Path
```

```python
participants_tsv = 'dummy_caps/subjects.tsv'
caps_directory = 'dummy_caps'

sub_ses_list = pd.read_csv(participants_tsv, sep='\t')
sub_ses_list
```

```python
for sub, ses in sub_ses_list[['participant_id', 'session_id']].values:
    print(sub, ses)
    pet = torch.randn(1, 28, 28)
    t1 = torch.randn(1, 28, 28)
    pet_linear_dir = Path(
        "dummy_caps", 
        "subjects", 
        sub, 
        ses, 
        "deeplearning_prepare_data", 
        "image_based", 
        "pet_linear"
    )
    if not os.path.exists(pet_linear_dir):
        os.makedirs(pet_linear_dir)
    t1_linear_dir = Path(
        "dummy_caps", 
        "subjects", 
        sub, 
        ses, 
        "deeplearning_prepare_data", 
        "image_based", 
        "t1_linear"
    )
    if not os.path.exists(t1_linear_dir):
        os.makedirs(t1_linear_dir)
    torch.save(pet, pet_linear_dir / f'{sub}_{ses}_pet_linear.pt')
    torch.save(t1, t1_linear_dir / f'{sub}_{ses}_t1_linear.pt')
```
